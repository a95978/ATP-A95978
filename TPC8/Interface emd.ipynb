```python
import PySimpleGUI as sg

def getEMD(linha):
    emd=[]
    novaLinha=linha.replace("\n","")
    campos=novaLinha.split(",")
    emd.append("emd"+str(int(campos[1])+1))
    for i in range(2,len(campos)):
            emd.append(campos[i])
    return emd
def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline() #retirar a linha de cabeçalho
    for linha in f:
        emd=getEMD(linha)
        bd.append(emd)
    return bd
BD=[]
BD=lerDataset('emd.csv')

def chaveOrd1(a):
    return a[7]

def modalidades(bd):
    bd.sort(key=chaveOrd1)
    L=[]
    for emd in bd:
        if(emd[7] not in L):
            L.append(emd[7])
    return L

def chaveOrd(a):
    data=a[1]
    data=data.split("-")
    return data

def listarDataset(bd):
    bd.sort(key=chaveOrd,reverse=True)
    L=[]
    print("id           Data           Nome        Apto")
    print("----------------------------------------------")
    for i in bd:
        L.append(i[0]+ " | "+i[1]+ " | "+i[3]+" "+i[2]+" | "+i[11])
    return L


def distribPorAno(bd):
    modalidade={}
    for a in bd:
        data=a[1]
        data=data.split("-")
        if data[0] in modalidade.keys():
            modalidade[data[0]]=modalidade[data[0]]+1
        else:
            modalidade[data[0]]=1
    return modalidade

def distribPorModalidade(bd):
    modalidade={}
    for a in bd:
        if a[7] in modalidade.keys():
            modalidade[a[7]]=modalidade[a[7]]+1
        else:
            modalidade[a[7]]=1
    return modalidade

def distrib(bd,i):
    modalidade={}
    for a in bd:
        if a[i] in modalidade.keys():
            modalidade[a[i]]=modalidade[a[i]]+1
        else:
            modalidade[a[i]]=1
    return modalidade

import matplotlib.pyplot as plt
def plotDistribPorModalidade(bd):
    d=distribPorModalidade(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.xlabel("Modalidade")
    plt.ylabel("Nº de alunos")
    plt.title("Distribuição por modalidades")
    plt.bar(x,y,align='center', width=0.8)
    plt.show()

def plotDistribAno(bd):
    d=distribPorAno(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.title("Distribuição Por Ano")
    plt.xlabel("Ano")
    plt.ylabel("Nº de alunos")
    plt.bar(x,y)
    plt.show()

def plotDistribClube(bd):
    d=distrib(bd,8)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.title("Distribuição por Clube")
    plt.xlabel("Clubes")
    plt.ylabel("Nº de alunos")
    plt.bar(x,y)
    plt.show()


menu_list_column =[
    [sg.Button("Carregar BD")],
    [sg.Button("Listar BD")],
    [sg.Button("Modalidades")],
    [sg.Button("Distribuição por ano")],
    [sg.Button("Distribuição por clube")],
    [sg.Button("Distribuição por modalidade")],    
]

data_viewer_column=[[sg.Text("Dados")],
    [sg.Text(size=(40, 10), key="_tarefas")]]

layout =[
    [
        sg.Column(menu_list_column),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]


window = sg.Window("Emd", layout)

stop = False
while not stop:
    event, values = window.read()
    if event == sg.WIN_CLOSED:
        stop = True
    elif event == "Carregar BD":
        window["_tarefas"].Update("BD foi carregada.")
    elif event == "Modalidades":
        window["_tarefas"].Update(modalidades(BD))
    elif event == "Listar BD":

        window["_tarefas"].Update(listarDataset(BD))
    elif event == "Distribuição por ano":
        window["_tarefas"].Update(plotDistribAno(BD))
    elif event == "Distribuição por clube":
        window["_tarefas"].Update(plotDistribClube(BD))
    elif event == "Distribuição por modalidade":
        window["_tarefas"].Update(plotDistribPorModalidade(BD))
    else:
        print(event + "\n")
        print(values)

window.close()
```
