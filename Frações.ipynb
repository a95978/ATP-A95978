```python
def menu():
    print("""Gestão de frações:
(1)Criar fração;
(2)Simplificar frações;
(3)Somar;
(4)Subtrair;
(5)Multiplicar;
(6)Dividir;
(7)Gerar lista de n frações;
(8)Somar lista;
(9)Maior fração;
(0)Sair""")
    
def criarFracao():
    numerador=int(input("Escreva o numerador: "))
    denominador=int(input("Escreva o denominador: "))
    return (numerador, denominador)

def criarFracao1(numerador, denominador):
    return (numerador, denominador)

def verFracao(f):
    return str(f[0]) + "/" + str(f[1])
    
def mdc(a,b): #Máximo divisor comum
    if(a % b==0):
        return b
    elif(b % a==0):
        return a
    else:
        if(a>b):
            return mdc(b, a % b)
        else:
            return mdc(a, b % a)
    
def simpleFracao(f): #Criar o tuplo
    num, denom = f
    a=mdc(num, denom)
    return (num//a, denom//a)

def somaFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simpleFracao(criarFracao1(n1*d2+n2*d1, d1*d2))

def subFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simpleFracao(criarFracao1(n1*d2-n2*d1, d1*d2))
        
def multiFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simpleFracao(criarFracao1(n1*n2,d1*d2))

def diviFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simpleFracao(criarFracao1(n1*d2,n2*d1))

def lista():
    from random import randint
    import random
    N=int(input("Introduza o nº de inteiros da lista:"))
    A=[]
    i=0
    for i in range(N):
        A.append(tuple(criarFracao1(random.randint(1,10),random.randint(1,10))))
    return A
    
def somalista(lista1):
    n1, d1=lista1[0]
    for i in range(1,len(lista1)):
        n2, d2=lista1[i]
        soma=criarFracao1(n1*d2+n2*d1, d1*d2)
        n1, d1=soma
    return simpleFracao(soma)
        
def maiorLista(lista1):
    maior=lista1[0]
    for i in range(1,len(lista1)):
        num, denom=lista1[i]
        maiorNum, maiorDenom = maior
        if((num/denom)>(maiorNum/maiorDenom)):
            maior=num,denom
    return maior
        
def opcoes():
    op="1"
    lis=[]
    while (op!="0"):
        menu()
        op=input("Qual é a sua opção?")        
        if(op=="1"):
            lis=criarFracao()
            print(verFracao(lis))
        elif(op=="2"):
            lis=simpleFracao(criarFracao())
            print(lis)
        elif(op=="3"):
            lis1=criarFracao()
            lis=criarFracao()
            lis2=somaFrac(lis,lis1)
            print(verFracao(lis2))
        elif(op=="4"):
            lis1=criarFracao()
            lis=criarFracao()
            lis2=subFrac(lis,lis1)
            print(verFracao(lis2))
        elif(op=="5"):
            lis1=criarFracao()
            lis=criarFracao()
            lis2=multiFrac(lis,lis1)
            print(verFracao(lis2))
        elif(op=="5"):
            lis1=criarFracao()
            lis=criarFracao()
            lis2=multiFrac(lis,lis1)
            print(verFracao(lis2))
        elif(op=="6"):
            lis1=criarFracao()
            lis=criarFracao()
            lis2=diviFrac(lis,lis1)
            print(verFracao(lis2))
        elif(op=="7"):
            lis2=lista()
            print(lis2)
        elif(op=="8"):
            lis=lista()
            print(lis)
            lis2=somalista(lis)
            print(verFracao(lis2))
        elif(op=="9"):
            lis=lista()
            print(lis)
            lis1=maiorLista(lis)
            print("O maior elemento é:", verFracao(lis1))
        elif(op=="0"):
            print("Até á próxima...")
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    opcoes()    
```
