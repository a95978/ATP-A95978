```python
def menu():
    print("""Gestão de listas de inteiros
(1)Introduzir a lista á mão;
(2)Gerar lista c/ valores aleatórios;
(3)O maior da lista;
(4)Ordenar a lista
(0)Sair""")
    
def criarListaM():
    N=int(input("Introduza o nº de inteiros da lista:"))
    L=[]
    i=0
    for i in range (N):
        elementos=input("Introduza o elemento "+str(i+1)+ ":")
        L.append(elementos)
    return L

def criarListaA():
    from random import randint
    import random
    N=int(input("Introduza o nº de inteiros da lista:"))
    A=[]
    i=0
    for i in range(N):
        A.append(random.randint(0,100))
    return A

def maiorElemento(lista):
    com=len(lista)
    maxi=lista[0]
    for i in range(com):
        if(lista[i]>maxi):
            maxi=lista[i]
    print("O máximo é:", maxi)
    
def bubbleSort(lista):
    com=len(lista)
    i=0
    j=0
    for j in range(com):
        for i in range(com-1):
            if(lista[i]>lista[i+1]):
                lista[i],lista[i+1],=lista[i+1],lista[i]
    return(lista)
    
def opcoes():
    op="1"
    lis=[]
    while (op!="0"):
        menu()
        op=input("Qual é a sua opção?")        
        if(op=="1"):
            lis=criarListaM()
            print(lis)
        elif(op=="2"):
            lis=criarListaA()
            print(lis)
        elif(op=="3"):
            if(len(lis)!=0):
                lis1=maiorElemento(lis)
                print(lis1)
            else:
                print("Escolha primeiro uma lista para criar")
                menu()
        elif(op=="4"):
            if(len(lis)!=0):
                lis2=bubbleSort(lis)
                print(lis2)
            else:
                print("Escolha primeiro uma lista para criar")
                menu()
        elif(op=="0"):
            print("Até á próxima...")
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    opcoes()    
```
