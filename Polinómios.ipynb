```python
def menu():
    print("""Gestão de polinómios:
(1)Criar polinómio;
(2)Calcular polinómio;
(3)Calcular tabela;
(4)Simplificar o polinómio;
(5)Calcular a derivada do polinómio;
(0)Sair""")
    
def criarTermo():
    coef=int(input("Introduza o coeficiente: "))
    exp=int(input("Introduza o expoente: "))
    return(coef, exp)


def inserirTermo(p,t):
    p.append(t)
    return p

def verTermo(t):
    c,e=t
    if e==0:
        return str(c)
    else:
        return (str(c) +"x^"+ str(e))
    
def verPolinomio(p):
    res=""
    for t in p:
        if res=="":
            res=verTermo(t)
        else:
            res=res + "+" + verTermo(t)
    print (res)
    
def calcPol(p,x):
    resultado=0
    for (c,e) in (p):
        resultado= resultado +c*(x**e)
    return (resultado)

def tabela(p,nlinhas):
    for i in range(nlinhas):
        print(str(i)+ "::"+ str(calcPol(p,i)))
        
def chaveOrd(t):
    (c,e)=t
    return e

def ordenarPol(p):
    p.sort(reverse=True, key=chaveOrd)
    
def maiorLista(lista):
    maiorG =lista[0][1]
    for i in range(1,len(lista)):
        maiorG1 =lista[i][1]
        if(maiorG1>maiorG):
            maiorG=maiorG1
    return maiorG
    
def simplePol(p,maiorG):
    L=[]
    i=maiorG
    while i>=0:
        resultado = 0
        for t in p:
            a1,a2=t
            if(a2==i):
                resultado=resultado+a1
        if(resultado>0): L.append((resultado,i))
        i=i-1
    return L

def derivarPol(p):
    res=[]
    for (c,e) in p:
        if e>0:
            res.append((c*e,e-1))
    return res

def opcoes():
    op="1"
    lis=[]
    while (op!="0"):
        menu()
        op=input("Qual é a sua opção?")        
        if(op=="1"):
            a=int(input("Introduza o nº de termos do polinómio: "))
            for i in range(a):
                lis1=inserirTermo(lis,criarTermo())
            print(verPolinomio(lis1))
        elif(op=="2"):
            x1=int(input("Introduza o valor de x: "))
            lis2=calcPol(lis,x1)
            print(lis2)
        elif(op=="3"):
            n=int(input("Introduza o nº de linhas da tabela: "))
            lis2=tabela(lis,n)
            print(lis2)
        elif(op=="4"):
            lis2=simplePol(lis,maiorLista(lis))
            print(verPolinomio(lis2))
        elif(op=="5"):
            lis2=derivarPol(lis)
            print(verPolinomio(lis2))
        elif(op=="0"):
            print("Até á próxima...")
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    opcoes()    
```
