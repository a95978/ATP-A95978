```python
import io
import matplotlib.pyplot as plt

#Exercício 1
def convInt(s):
    return int(s)

import io
def lerDataset(fnome):
    f=io.open(fnome, encoding="utf-8")
    bd=[]
    f.readline() #le só a primeira linha #Sempre que faço o f vai aumentado como um apontador)
    for linha in f:
        bd.append(getAluno(linha))
    return bd

#Tirar as aspas utilizo o replace
#Strip é para espaços
def getAluno(texto):
    textoLimpo=texto.replace("\"","")
    campos=texto.split(",")
    aluno=[]
    aluno.append(campos[0])
    aluno.append(campos[1])
    aluno.append(campos[2])
    aluno.append(campos[3:])
    Lista=[]
    for s in aluno[3]: #Converter os tpcs em inteiros
        Lista.append(int(s))
    aluno[3]=Lista
    return aluno
BD=[]
BD=lerDataset("alunos.csv") #Depende de onde estás, tens de por o percurso
print(BD) #.. é re ferencia pra .py 

#Exercício 2
def chaveOrd(a):
    return a[1]
    
def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print("id     nome     curso     média")
    print("-------------------------------")
    for a in bd:
        print(a[0]+"|"+a[1]+"|"+a[2]+"|"+ str(sum(a[3])/4))
BD=[]
BD=lerDataset("alunos.csv")

#Exercício 3
def identificacao():
    id=input("Introduza o id do aluno com aspas: ")
    return id
def consultarDataset(bd, id):
    L=[]
    for a in bd:
        if(a[0]==id):
            L.append(a[0])
            L.append(a[1])
            L.append(a[2])
            L.append(sum(a[3])/4)
    print(L)
BD=[]
BD=lerDataset("alunos.csv")
consultarDataset(BD,identificacao())

#Exercício 4
def chaveOrd(a):
    a[3]=sum(a[3])/4
    return a[3]
def top10(bd):
    bd.sort(key=chaveOrd)
    bd.reverse()
    i = 0
    for a in bd:
        if(i<10):
            print(a)
        i=i+1
BD=[]
BD=lerDataset("alunos.csv")
top10(BD)

#Exercício 5
def distribPorCurso(bd):
    distribuicao={}
    for a in bd: 
        if a[2] in distribuicao.keys():
            distribuicao[a[2]]=distribuicao[a[2]]+1
        else:
            distribuicao[a[2]]=1
    return distribuicao
BD=[]
BD=lerDataset("alunos.csv")
d=distribPorCurso(BD)
print(d)

#Exercício 6
def distribPorMedia(bd):
    distribuicao={}
    for a in bd:
        a[3]=sum(a[3])/4
        if round(a[3]) in distribuicao.keys():
            distribuicao[round(a[3])]=distribuicao[round(a[3])]+1
        else:
            distribuicao[round(a[3])]=1
    return distribuicao
BD=[]
BD=lerDataset("alunos.csv")
d=distribPorMedia(BD)
print(d)

#Exercício 7
def plotDistribPorCurso(bd):
    d=distribPorCurso(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.xlabel("Curso")
    plt.ylabel("Nº de alunos")
    plt.title("Distribuição por curso")
    plt.bar(x,y)
    plt.show()
BD=[]
BD=lerDataset("alunos.csv")
plotDistribPorCurso(BD)

#Exercício 8
def plotDistribPorMedia(bd):
    d=distribPorMedia(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.xlabel("Média")
    plt.ylabel("Nº de alunos")
    plt.title("Distribuição por média")
    plt.bar(x,y)
    plt.show()
BD=[]
BD=lerDataset("alunos.csv")
plotDistribPorMedia(BD)
```
